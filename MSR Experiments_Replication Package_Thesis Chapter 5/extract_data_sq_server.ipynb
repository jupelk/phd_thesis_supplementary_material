```python
# Step 4: read main data CSV file and add code smells and code smells diff data to it

import pandas as pd
new_data_df = pd.read_csv('c:\\CodeBook\\new_code_rework_TD\\sample_main_rows_with_aggregated_data_and_sq_code_smells.csv')
new_main_rows = new_data_df[new_data_df['hash'].notna()]
new_main_rows['code_smells'] = 0
new_main_rows['code_smells_diff'] = 0
for i, row in new_main_rows.iterrows():
    new_main_rows.at[i,'code_smells'] = int(json_data["measures"][0]["history"][int(i)]["value"])
    if i > 0:
        new_main_rows.at[i,'code_smells_diff'] = new_main_rows.at[i,'code_smells'] - new_main_rows.at[i-1,'code_smells']
    #print(int(json_data["measures"][0]["history"][int(i)]["value"]))
    print (new_main_rows.at[i,'code_smells'], new_main_rows.at[i,'code_smells_diff'])

# Step 5: write new dataframe to a new CSV file
new_main_rows.to_csv('c:\\CodeBook\\new_code_rework_TD\\S_sample_main_rows_sq_code_smells.csv', index=False)
```
