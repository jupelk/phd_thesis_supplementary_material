# Types of Veracity Requirements Supported in Software Solutions Developed by Practitioners

```python
df['Q23'].dropna()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation for demonstration (replace this with your actual DataFrame)
# df = pd.DataFrame({'Q23': ['Regulatory Veracity requirements', 'Data Veracity requirements', 'Financial Data Veracity requirements', ...]})

#count number of responses for regulatory veracity
reg_v = df['Q23'].str.contains('Regulatory Veracity requirements').sum()
reg_v_percentage = (reg_v/len(df['Q23'].dropna()))*100

#count number of responses for data veracity
data_v = df['Q23'].str.contains('Data Veracity requirements').sum()
data_v_percentage = (data_v/len(df['Q23'].dropna()))*100

#count number of responses for financial data veracity
financial_data_v = df['Q23'].str.contains('Financial Data Veracity requirements').sum()
financial_data_v_percentage = (financial_data_v/len(df['Q23'].dropna()))*100

#count number of responses for process veracity
process_v = df['Q23'].str.contains('Process Veracity requirements').sum()
process_v_percentage = (process_v/len(df['Q23'].dropna()))*100

#count number of responses for cultural veracity
cultural_v = df['Q23'].str.contains('Cultural Veracity requirements').sum()
cultural_v_percentage = (cultural_v/len(df['Q23'].dropna()))*100

#count number of responses for other veracity
other_v = df['Q23'].str.contains('Other').sum()
other_v_percentage = (other_v/len(df['Q23'].dropna()))*100

#plot all counts reg_v, data_v, financial_data_v, process_v, other_v

#change color of chart to grey
plt.rcParams['axes.facecolor'] = 'white'

percentages = [reg_v_percentage, data_v_percentage, financial_data_v_percentage, cultural_v_percentage, process_v_percentage, other_v_percentage]


#make labels 90 degrees

#plt.bar(['Regulatory', 'Data', 'Financial Data', 'Cultural', 'Process', 'Other'], [reg_v, data_v, financial_data_v, cultural_v, process_v, other_v], color='grey', width=0.6)
#percentages
bars = plt.bar(['Regulatory', 'Data', 'Financial Data', 'Cultural', 'Process', 'Other'], [reg_v_percentage, data_v_percentage, financial_data_v_percentage, cultural_v_percentage, process_v_percentage, other_v_percentage], color='grey', width=0.6)
plt.xlabel('Types of Veracity Requirements')
plt.ylabel('Percentage of Respondents who Voted')
plt.title('Types of Veracity Requirements Supported in Software Solutions Developed by Practitioners')
plt.xticks(rotation=90)

# Add percentage labels on top of each bar
for bar, percentage in zip(bars, percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage:.1f}%', ha='center', va='bottom')


#size of figure
plt.figure(figsize=(12,8))

#increase spacing between bars
plt.tight_layout()

plt.show()

#save the plot
plt.savefig('/Users/jupelk/Documents/PhD Experiments/Survey/Plots/VR_in_Apps.png')



# SF: Might be interesting. No cultural veracity?
```

# Veracity Requirement Types most useful to quantify Costs, Benefits, and Consequences

```python
# Assuming 'df' is your DataFrame
df_long = df.melt(value_vars=['Q49', 'Q50.1', 'Q51'], var_name='Question', value_name='Category')

df_long['Category'].replace('Other (Please specify the type and why)', 'Other', inplace=True)

# Here df_long will have two columns: 'Question' indicating the question number and 'Category' the response.
import seaborn as sns
import matplotlib.pyplot as plt

# Mapping of question codes to descriptive names
question_names = {
    'Q49': 'Cost to fix',
    'Q50.1': 'Benefit of fixing',
    'Q51': 'Consequences of not fixing',
}

# Define the category order
category_order = ['Data', 'Financial Data', 'Regulatory', 'Process', 'Cultural', 'Other']

plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df_long, x='Category', hue='Question', palette='Set2', order=category_order)

# Rotate x-axis labels 90 degrees
plt.xticks(rotation=90)

# Title and labels
plt.title('Veracity Requirement types most useful to quantify Costs, Benefits and Consequences')
plt.xlabel('Type of Veracity Requirements')
plt.ylabel('Number of respondents who voted')
plt.legend(title='Concepts', labels=[question_names[q] for q in df_long['Question'].unique()])

# Add counts on the bars
for p in ax.patches:
    count = int(p.get_height())
    if count > 0:  # Only annotate non-zero bars
        ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), 
                    textcoords='offset points')

plt.tight_layout()  # Adjust layout for better readability
plt.show()
```
